```python
# clean up notes / exceptions
# comedy central presents _____ , the standups _____, 3 name comics ( lous c k , roy wood jr)
# but most are first two words after the date, parse by '-'
com_test = df2.iloc[0,0]
com_test = com_test.split('/')[-2]
print(com_test)
com_test = com_test.split('-')[0:2]
print(com_test)
comic = com_test[0] + ' ' + com_test[1]
print(comic)

def comic_extract(com):
    com = com.split('/')[-2]
    com = com.split('-')[0:2]
    com = com[0] + ' ' + com[1]
    return com
```

```python
df3 = df2.copy()
df3.comedian.astype(str)
df3.comedian = df3.comedian.apply(comic_extract)
```

```python
# Now clean up transcripts!
df5 = df4.copy()
print(df5.iloc[2,1].find('RELATED POSTS'))
end_trim = df5.iloc[2,1].find('RELATED POSTS')
df5.iloc[2,1] = df5.iloc[2,1][0:end_trim]
```

```python
def transcript_trim(script):
    end_trim = script.find('RELATED POSTS')
    script = script[0:end_trim]
    return script
```

```python
df6 = df5.copy()
df6.transcript = df6.transcript.astype(str)
df6.transcript = df6.transcript.apply(transcript_trim)
```

```python
# now concatenate all transcripts for a given comedian
#df.groupby('A').agg({'B':'sum','C':'-'.join})
df7 = df6.copy()
print(df7.shape)
df7 = df7.groupby('comedian').agg({'transcript':' '.join}).reset_index()
df7.iloc[34,0] = 'dave chappelle'
df7.iloc[31,0] = 'dave attell'
df7.iloc[51,0] = 'gabriel iglesias'
df7.iloc[119,0] = 'sarah silverman'
print(df7.iloc[119,0])
df7 = df7.groupby('comedian').agg({'transcript':' '.join}).reset_index()
print(df7.shape)
# misspelled names: david chappelles, dave attel, gabriel fluffy, sarah silvermans
```
